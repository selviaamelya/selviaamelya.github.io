Buatlah sebuah program untuk memasukkan dan menghitung biaya laundry dan dry clean dengan beberapa ketentuan sebagai berikut:
1. Biaya laundry per-kg Rp. 6000 dan dry clean per-item Rp. 35000.
2. Terdapat diskon 5% jika melakukan laundry lebih dari 10 kg. Sedangkan dry clean juga diskon 5% jika melakukan dry clean lebih dari 5 item.
3. Terdapat diskon member 3% untuk member bronze, 4% untuk member silver, dan 5% untuk member gold. Diskon diterapkan pada total biaya akhir.

Selanjutnya, lengkapilah Tabel 2 berikut sesuai menggunakan program yang telah anda buat. Dan Selanjutnya hitunglah harga berikut:

1. (Nama: Vincent, Laundry: 11 Kg, Dry Clean: 6 item, Status member: Gold, Harga: ...Rp234600)
2. (Nama: Aldi, Laundry: 7 Kg, Dry Clean: 8 item, Status member: Silver, Harga: ...Rp293020)
3. (Nama: Dewi, Laundry: 3 Kg, Dry Clean: 2 item, Status member: Non-member, Harga: ...Rp88000)


```python
#Case 2 - Versi diskon di akhir
name = input("Tuliskan nama Anda: ")
laundry = int(input("Berapa kg berat laundry Anda? "))
dry_clean = int(input("Berapa item jumlah dry clean Anda? "))
membership = input("Apa jenis member Anda? Tulis 'b' untuk member Bronze, 's' untuk Silver,'g' untuk member Gold, dan 'n' jika tanpa member: ")

harga_awal = (int(laundry) * 6000) + (int(dry_clean) * 35000)
diskon = 0
diskon_laundry = 0
diskon_dry_clean = 0
diskon_membership = 0
if laundry > 10:
  diskon_laundry = (5/100) * harga_awal
if dry_clean > 5:
  diskon_dry_clean = (5/100) * harga_awal

if membership == "b":
   diskon_membership = (3/100) * harga_awal
if membership == "s":
   diskon_membership = (4/100) * harga_awal
if membership == "g":
   diskon_membership = (5/100) * harga_awal

diskon += diskon_laundry + diskon_dry_clean + diskon_membership
harga_akhir = round(harga_awal - diskon)

print(f"Total harga untuk laundry dan dry clean Anda adalah Rp{round(harga_akhir)}.")
```

Buatlah sebuah program untuk menghitung gaji yang harus diterima oleh karyawan dengan ketentuan sebagai berikut:
-	Gaji pokok: Golongan III.a Rp2775000 dan Golongan III.b Rp. 3000000
-	Jika sudah berkeluarga maka terdapat tunjangan: Istri Rp. 250000, Anak Rp. 250000 untuk setiap anak yang dimiliki (maksimal 2 anak)
-	Uang kehadiran: Rp37500 per hari
- Uang makan: Rp35000 per hari (per kehadiran)

Kemudian hitunglah:
- Nama: Andi; Golongan: 3a; Sudah berkeluarga; 3 Anak; 19 hari kehadiran; Gaji yang diterima ...Rp4902500
- Nama: Andre; Golongan: 3b; Belum berkeluarga; 0 Anak; 20 hari kehadiran; Gaji yang diterima ...Rp4450000.

```python
#Case 3
name = input("Masukkan nama: ")
golongan = input("Masukkan jenis golongan, 3a atau 3b: ")
berkeluarga = input("Apakah sudah berkeluarga? y atau t: ")
if berkeluarga == "y":
  anak = int(input("Apakah memiliki anak? Berapa jumlah anak? Jika tidak ada isi '0': "))
hari_hadir = int(input("Berapa hari kehadiran? "))

gaji = 0
#cek golongan
if golongan == "3a":
  gaji += 2775000
elif golongan == "3b":
  gaji += 3000000

#cek status berkeluarga dan jumlah anak
if berkeluarga == "y":
  gaji += 250000
  if anak == 1:
    gaji += 250000
  elif anak > 1:
    gaji += 250000 * 2
  else:
    gaji += 0
else:
  gaji += 0

#tambahan gaji berdasarkan jumlah hari kehadiran
uang_kehadiran_makan = hari_hadir * (37500 + 35000)
gaji += uang_kehadiran_makan

print(f"Gaji yang diterima {name} adalah Rp{gaji}.")
```

Dengan menggunakan perulangan while atau for buatlah:
  1. Fungsi untuk menampilkan bilangan 1 sampai 200 yang habis dibagi dengan 7 dan merupakan bilangan genap
  2. Fungsi untuk menampilkan deret 0, 1, 1, 2, 3, 5, ..., 4181

```python
#Case 4 - Fungsi 1
def fungsi_pertama (bilangan):
  for angka in bilangan:
    if angka % 7 == 0 and angka % 2 == 0:
      print(angka)

bilangan = range(1, 201)
fungsi_pertama(bilangan)
```

Buatlah sebuah fungsi untuk menampilkan bilangan prima hingga suku ke n:

```python
#Case 5
def prima(n):
  bilangan = range(2,n)
  for angka in bilangan:
    if angka == 2 or angka == 3 or angka == 5 or angka == 7 or angka == 11:
      print(angka)
    elif angka % 2 == 0:
        continue
    elif angka % 3 != 0 and angka % 5 != 0 and angka % 7 != 0:
      print(angka)

prima(100)
```
